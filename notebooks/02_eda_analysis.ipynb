```python
!pip install seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load cleaned data
df = pd.read_csv( r"E:\Hospitality_Analytics_Project\data\processed\cleaned_hotel_booking_dataset.csv")


#1. Cancellation Rate
sns.countplot(
    x='is_canceled',
    data=df
)

plt.show()


#2. Lead Time vs Cancellation
sns.boxplot(
    x='is_canceled',
    y='lead_time',
    data=df
)

plt.show()


#3. ADR vs Cancellation
sns.boxplot(
    x='is_canceled',
    y='adr',
    data=df
)

plt.show()


#4. Market Segment Analysis
df['market_segment'].value_counts()
sns.countplot(
    y='market_segment',
    data=df
)

plt.show()


#5. Country Analysis
df['country'].value_counts().head(10)


#6. Monthly Booking Trend
sns.countplot(
    x='arrival_date_month',
    data=df
)

plt.xticks(rotation=90)

plt.show()


#7. Correlation Heatmap
sns.heatmap(
    df.corr(numeric_only=True)
)

plt.show()
```
