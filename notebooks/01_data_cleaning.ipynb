```python
%pip install pandas numpy

import pandas as pd
import numpy as np


#Load the dataset
df = pd.read_csv(r"E:\Hospitality_Analytics_Project\data\raw\hotel_booking.csv")
df.head()


#Check the dataset
print(df.shape)
df.info()
df.isnull().sum()


#Remove PII Columns
df.drop(
    columns=[
        'name',
        'email',
        'phone-number',
        'credit_card'
    ],
    inplace=True
)


#Handle Missing Values
df['children'].fillna(0,inplace=True)

df['country'].fillna(
    df['country'].mode()[0],
    inplace=True
)

df['agent'].fillna(0,inplace=True)

df['company'].fillna(0,inplace=True)


#Remove Duplicates
df.drop_duplicates(inplace=True)


#Feature Engineering
#1. Total Guests
df['total_guests'] = (
    df['adults']
    + df['children']
    + df['babies']
)

#2. Total Stay
df['total_stay'] = (
    df['stays_in_weekend_nights']
    +
    df['stays_in_week_nights']
)

#3. Total Revenue
df['total_revenue'] = (
    df['adr']
    *
    df['total_stay']
)


#Save the cleaned dataset
df.to_csv(
    r"E:\Hospitality_Analytics_Project\data\processed\cleaned_hotel_booking_dataset.csv",
    index=False
)

print("Data cleaning and feature engineering completed. Cleaned dataset saved successfully.")
```
